Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom.

```python
#%% 1 - IMPORTAR A BASE DE DADOS
```

```python
import pandas as pd

df = pd.read_csv('clientes.csv')
display(df)
```

```python
#%% 2 - VISUALIZAR A BASE DE DADOS
display(df.info())
```

```python
#%% 3 - TRATAMENTO DOS DADOS
```

```python
df = df.drop(columns='id_cliente')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
codificador_profissao = LabelEncoder()
df['profissao'] = codificador_profissao.fit_transform(df['profissao'])

codificador_mix = LabelEncoder()
df['mix_credito'] = codificador_mix.fit_transform(df['mix_credito'])

codificador_comportamento = LabelEncoder()
df['comportamento_pagamento'] = codificador_comportamento.fit_transform(df['comportamento_pagamento'])

display(df)
```

```python
#%% 4. TREINAMENTO DA IA
y = df['score_credito']
x = df.drop(columns='score_credito')
```

```python
#visualizando as colunas para observar se ha linhas vazias, se tem algum tipo de dado a ser tratado
display(df.info())
```

```python
# campos de treino e de teste
from sklearn.model_selection import train_test_split

# test_size=0.3 -> esse parametro indica que 30% dos dados serao separados para teste e os 70% serao para treino
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
# modelo de previsao -> se score esta bom, ruim ou ok
# 1 - importar a IA (modelo)
#modelo 1
from sklearn.ensemble import RandomForestClassifier
#modelo 2
from sklearn.neighbors import KNeighborsClassifier

# 2 - criar a IA (modelo)
#modelo 1
modelo_arvoredecisao = RandomForestClassifier()
#modelo 2
modelo_knn = KNeighborsClassifier()

# 3 - treinar a IA (modelo)
#modelo 1
modelo_arvoredecisao.fit(x_treino, y_treino)
#modelo 2
modelo_knn.fit(x_treino, y_treino)
```

```python
#%% 5. TESTE PARA ESCOLHA DO MELHOR MODELO DE PREVISAO
```

```python
from sklearn.metrics import accuracy_score
# Avaliar e escolher o melhor modelo de previsao

#modelo 1
previsao_arvore = modelo_arvoredecisao.predict(x_teste)
#modelo 2
previsao_knn = modelo_knn.predict(x_teste)

# comparar a previsao com o y_teste
from sklearn.metrics import accuracy_score

display(accuracy_score(y_teste, previsao_arvore))
display(accuracy_score(y_teste, previsao_knn))
```
